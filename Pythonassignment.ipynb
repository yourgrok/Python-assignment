```python
aman = {"maths": 66, "hindi": 88, "english": 65, "chem": 90, "physics": 54}
salman = {"maths": 78, "hindi": 59, "english": 87, "chem": 78, "physics": 65}
rishi = {"maths": 78, "hindi": 67, "english": 95, "chem": 70, "physics": 59}
hemant = {"maths": 78, "hindi": 69, "english": 63, "chem": 57, "physics": 78}


amanavg = sum(aman.values())/5
salmanavg = sum(salman.values())/5
rishiavg = sum(rishi.values())/5
hemantavg = sum(hemant.values())/5

maxavg = max(amanavg,salmanavg,rishiavg,hemantavg)
print(maxavg)
if maxavg == amanavg:
  print("aman")
elif maxavg == salmanavg:
  print("salman")
elif maxavg == rishiavg:
  print("rishi")
else:
  print("hemant")

```
